```python
#traing the input and output values
from sklearn.model_selection import train_test_split
feature_columns = ['age','weight','hoursRunInAWeek','highHealWear','multipleJointPains','allignmentOfMidFoot','diabetes','cyanosis','burningSensationInPS','Tenderness']
predicted_class = ['Status']
X = data[feature_columns].values
y = data[predicted_class].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state=10)
```

```python
print("number of rows missing age: {0}".format(len(data.loc[data['age'] == 0])))
print("number of rows missing weight: {0}".format(len(data.loc[data['weight'] == 0])))
print("number of rows missing hoursRunInAWeek: {0}".format(len(data.loc[data['hoursRunInAWeek'] == 0])))
print("number of rows missing highHealWear: {0}".format(len(data.loc[data['highHealWear'] == 0])))
print("number of rows missing multipleJointPains: {0}".format(len(data.loc[data['multipleJointPains'] == 0])))
print("number of rows missing allignmentOfMidFoot: {0}".format(len(data.loc[data['allignmentOfMidFoot'] == 0])))
print("number of rows missing diabetes: {0}".format(len(data.loc[data['diabetes'] == 0])))
print("number of rows missing cyanosis: {0}".format(len(data.loc[data['cyanosis'] == 0])))
print("number of rows missing burningSensationInPS: {0}".format(len(data.loc[data['burningSensationInPS'] == 0])))
print("number of rows missing Tenderness: {0}".format(len(data.loc[data['Tenderness'] == 0])))
```

```python
#converting the zeros into mean values for the prensented rows and column
from sklearn.preprocessing import Imputer
fill_values = Imputer(missing_values=0, strategy="mean", axis=0)
X_train = fill_values.fit_transform(X_train)
X_test = fill_values.fit_transform(X_test)
```

```python
#applying algorithm for trained datas
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train,y_train)
```

```python
predict_train_data = gnb.predict(X_test)
predict_train_data
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predict_train_data))
```

```python
#creating module using pickle
import pickle
with open("module.pickle","wb") as f:
    pickle.dump(gnb,f)
with open("module.pickle","rb") as f:
    model=pickle.load(f)
```
